```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from waveoptics.datasets.loaders import load_matfile
from waveoptics.tensors.stitch import stitch_array
from waveoptics.plots.plots import complex_imshow
```

```python
tm = load_matfile('../resources/matrice_34_34_exp.mat')['tm_corr']
weights = load_matfile('../resources/weights.mat')
```

```python
stitched_array = stitch_array(tm, crop_outputs=True, cropped_length=180)
stitched_array = stitched_array / np.nanmax(np.abs(stitched_array))

if np.iscomplexobj(stitched_array):
    fig, ax = complex_imshow(stitched_array)
else:
    plt.imshow(stitched_array, cmap='gray')
```
