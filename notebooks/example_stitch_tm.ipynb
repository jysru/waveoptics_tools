```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from waveoptics.datasets.loaders import load_matfile
from waveoptics.tensors.stitch import stitch_tm
from waveoptics.plots.plots import complex_imshow
```

```python
tm = load_matfile('../resources/matrice_34_34_exp.mat')['tm_corr']
weights = load_matfile('../resources/weights.mat')
```

```python
stitched_tm = stitch_tm(tm, weights, crop_outputs=True, cropped_length=180)
stitched_tm = stitched_tm / np.nanmax(np.abs(stitched_tm))

if np.iscomplexobj(stitched_tm):
    fig, ax = complex_imshow(stitched_tm)
else:
    plt.imshow(stitched_tm, cmap='gray')
```
