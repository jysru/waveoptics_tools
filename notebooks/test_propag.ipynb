```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from waveoptics.beams.numpy import gaussian_1d, gaussian_2d
from waveoptics.propag.numpy import fft_1d, fft_2d, frt_1d, frt_2d
```

```python
x = np.arange(-1e-3, +1e-3, 1e-5)

y = gaussian_1d(x, 1e-4)
z = frt_1d(y, dz=20e-3)

plt.plot(x, y)
plt.show()
plt.plot(np.abs(z))
```

```python
X, Y = np.meshgrid(x, x)
Y = gaussian_2d((X, Y), widths=(1e-4, 2e-4))
Z = frt_2d(Y, dz=50e-3)

plt.imshow(Y)
plt.show()

plt.imshow(np.abs(Z))
plt.show()
```
